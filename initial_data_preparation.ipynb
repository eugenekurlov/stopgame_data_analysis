# Initial data preparation

```python
# import libraries

import pandas as pd
import numpy as np
```

```python
# read the parsed data
df = pd.read_csv('stopgame.csv', sep='\t')
df
```

```python
# write function for converting russian string date format to standard datetime format

from datetime import date
def ruDateToDate(rudate):
    rudate = str(rudate)
    ru_month = ['января','февраля','марта','апреля','мая','июня','июля','августа','сентября','октября','ноября','декабря']
    dmg = rudate.split()
    if len(dmg)==3:
        if ((len(dmg[0])==1) | (len(dmg[0])==2)) & (dmg[0].isdigit()):
            day = int(dmg[0])
        else:
            print('A day should consist of number with two digits from 01 to 31')
        if (len(dmg[2])==4) & (dmg[2].isdigit()):
            year = int(dmg[2])
        else:
            print('A year should be a 4-digit number')
        if dmg[1] in ru_month:
            month = ru_month.index(dmg[1])+1
        else:
            print('The month must be written in Russian letters in the genitive case')
    else:
        print("Wrong date. The format 'dd month_with_Russian_letters yyyy' is needed ")
    normdate = date(year, month, day)
    numpy_date = np.datetime64(normdate)
    pd_date = pd.DatetimeIndex([normdate], dayfirst=True)[0]
    return numpy_date
```

```python
# some dates have unfull format. Drop the rows containing them
for index, row in df.iterrows():
    try:
        df.loc[index, 'Release_date'] = ruDateToDate(row['Release_date'])
    except UnboundLocalError: 
        df = df.drop(index)
```

There were only two games without specified release date, so we dropped corresponding rows.

```python
# reset index
df = df.reset_index(drop=True)
df
```

The views have values floored to thousands, which is denoted with letter 'K'. Designations '25K' beside two videos don't mean that they both have equal number like 25000. The 'K' symbol can be replaced with three-digir random number without loss of generality.

```python
# get initial info about dataset
df.info()
```

```python
# data cleaning


df.loc[df['Views'].str.contains('K'),'Views'] = (df
                                                 .loc[df['Views']
                                                 .str.contains('K'), 'Views']
                                                 .str.replace('K', '')
                                                 .astype(int)*1000) + \
                                                np.random.randint(100,1000, sum(df['Views'].str.contains('K')))
```

```python
# replace russian shortened names of months with their number in the year
ru_month_shortened = ['янв.','февр.','мар.','апр.','мая','июн.','июл.','авг.','сент.','окт.','нояб.','дек.']
ru_month_number = map(str, list(range(1, len(ru_month_shortened)+1)))
months = dict(zip(ru_month_shortened, ru_month_number))

df['Review_date'] = (df['Review_date'].replace(months, regex=True)
                        .str.replace(' ','.')) # regex=True is necessary
# concatenate these date with the current year (2023)
df.loc[df['Review_date'].str.count('\.') == 1, 'Review_date'] = df.loc[df['Review_date'].str.count('\.') == 1, 'Review_date']+'.23'

# convert the Release_date and Review_date to datetime format
df['Review_date'] = pd.to_datetime(df['Review_date'], dayfirst=True)
df['Release_date'] = pd.to_datetime(df['Release_date'])
```
